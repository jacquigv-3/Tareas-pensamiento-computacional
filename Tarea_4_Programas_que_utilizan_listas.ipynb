### 1. Analizador de una lista entera (todo-en-uno de una sola lista)

**Descripción**

Lee n>0, captura n enteros y realiza:

1. Muestra con índice creciente en formato lista[i] = valor.

2. Muestra con índices negativos en orden inverso: lista[-1] = ...

3. Calcula y muestra el promedio.

4. Cuenta PARES e IMPARES (0 es par).

5. Construye y muestra:

- la lista al cuadrado,

- la sublista de pares,

- la sublista de impares.


**Entradas**
- Un entero n (repetir lectura hasta que n>0).

- n enteros, uno por renglón.

**Salidas (en este orden exacto)**
1. lista[i] = valor para i de 0 a n-1

2. lista[-k] = valor para k de 1 a n

3. La lista en formato Python, p. ej. [6, 0, 1, 2, 4]

4. El promedio (como float) en renglón aparte

5. PARES=x y IMPARES=y (cada uno en su renglón)

6. CUADRADOS y la lista al cuadrado

7. PARES_LISTA y la sublista de pares

8. IMPARES_LISTA y la sublista de impares


```python
# El primer paso es hacer nuestra lista.
print("""Creación de lista:

""")
  #Se crea un while True para recibir el tamaño deseado por el
  # usuario de mi lista.

  # Leer un entero n mayor que 0 (n>0)
while True:
  n= int(input("Introduce un número entero n (>0): ")) #Longitud de la lista
  if n>0:
    break

  # Capturar n enteros
  ## Sintáxis de range (elem0, elemenF, incremento)
  ### El elemento 0 no es obligatorio porque si no se pone inica en 0.
  ### El incremento no es obligatorio, si no se pone incrementa de 1
  ### en 1.

lista = [] #Se crea una lista llamada "lista" en donde se recopilaran
           # los elementos.

for i in range(n): #Aqui hace el rango de la lista, si n=5, se harán
                   #5 elementos en la lista.
  num = int(input(f"Introduce el entero {i+1}: ")) #se pone "i+1"
# para que al usuario le aparezca en lugar del valor 0, valor 1
# (formato más user-friendly)
  lista.append(num) #Agrega el elemento ingresado por el usuario
  #hasta el final (se va recolectando)

print("""
Requerimiento #1: Muestra con índice creciente en formato lista[i] = valor"""
      )

#Mostrar con índice creciente
  # Imprimimos una cadena de texto formateada (f-string).
  # La 'f' al inicio nos permite insertar variables usando {}.

for i in range(n):
  print(f"lista{i} = {lista[i]} ") #las llaves indican que será el
  # índice, ya que con este se accede a la posición.

print("""
Requerimiento #2: Muestra con índices negativos en orden inverso: lista[-1] = ...""")

# Mostrar con índices negativos en orden inverso:lo que pasa con los
# índices neg es que muestran tus elementos del final al principio.

## Queremos que la variable 'k' tome los valores: 1, 2, 3, ...,
## hasta n -> Por eso usamos range(1, n + 1):
### Inicia en 1: Porque el primer índice negativo que nos interesa
### es -1.
###Termina en n+1: Porque range() se detiene un número ANTES del final.
###Para que el rango INCLUYA 'n', debemos poner 'n+1'.for k in range
##### (1,n+1):

for k in range (1, n+1):
  print(f"lista[-{k}] = {lista[-k]} ")

print("""
Requerimiento #3: La lista en formato Python""")
#Mostrar la lista completa
print(f"Lista completa: {lista} ")

print("""
Requerimiento #4: Calcula y muestra el promedio""")

#Calcular y mostrar el promedio
promedio = sum(lista)/n
print(promedio)

print("""
Requerimiento #5: PARES=x y IMPARES=y""")

#Contar pares e impares
  ### Lectura de condición: "Suma (sum) un '1' por cada (for) elemento 'x' en la lista,
  ### solo si (if) el residuo de 'x' dividido entre 2 es cero
   ## (x % 2 == 0)".
    # Sumar '1' por cada acierto es una forma de CONTAR.
pares = sum(1 for x in lista if x%2 == 0) # el 1 suma el valor si
## sus condiciones son verdaderas
impares = n - pares # cantidad de elementos - pares
                    #(valor calculado anteriormente)
print(f"PARES = {pares}")
print(f"IMPARES = {impares}")

print("""
Requerimiento #6: la lista al cuadrado""")

#Construir y mostrar la lista al cuadrado
  ## Crea una lista nueva poniendo el cuadrado (x**2)
  # de cada elemento (x) que está en lista".
cuadrados = [x**2 for x in lista]
print(f"CUADRADOS = {cuadrados}")

print("""
Requerimiento #7: sublista de pares""")

#Construir y mostrar la sublista de pares
  ## Crea una lista nueva (pares_lista) poniendo
  #  la x original, pero solo si (if) esa x es par (x % 2 = 0)".
pares_lista = [x for x in lista if x % 2 == 0]
print(f"PARES_LISTA = {pares_lista}")

print("""
Requerimiento #8: sublista de impares""")
#Construir y mostrar la sublista de impares
  ## Crea una lista nueva (impares_lista) poniendo
  # la x original, pero solo si (if) esa x es impar (x % 2 != 0)".
impares_lista = [x for x in lista if x % 2 != 0]
print(f"IMPARES_LISTA = {impares_lista}")
```

### 2. Dúo de listas: suma por posición, ordenación y sin duplicados

**Descripción**

Trabaja con dos listas de enteros del mismo tamaño.

1. Lee un entero m>0 (repetir hasta cumplir).

2. Captura m enteros para L1 y luego m enteros para L2.

3. Genera LSUMA donde LSUMA[i] = L1[i] + L2[i].

4. Crea LORDENADA concatenando L1+L2 y ordenando de menor a mayor.

5. Crea LUNICA eliminando duplicados de LORDENADA preservando el orden de aparición.

**Entradas**
- Un entero m>0.

- m enteros para L1.

- m enteros para L2.

**Salidas (en este orden)**
- L1= (en formato lista)

- L2=

- LSUMA=

- LORDENADA=

- LUNICA= (sin duplicados, orden estable)

```python
# El primer paso es hacer nuestra lista.

print("""Creación de listas:

""")

# Definir el tamaño de las listas
  #Se crea un while True para recibir el tamaño deseado por el
  #usuario de mi lista.
  #Leer un elemento m mayor que 0

while True:
  m = int(input("Introduce un número entero m (>0):"))
  if m > 0:
    break

print("Creación de lista #1: ")

# Capturar m enteros para L1

L1 = [] #Se crea una lista llamada "L1" en donde se recopilaran los
        #elementos.
for i in range(m):
  num = int(input(f"Introduce el entero {i+1} para L1: ")) #input
  # para que el usuario coloque los elementos deseados para L1.
  L1.append(num) #se van agregando s mi L1 los números "num"
print(L1)

print("Creación de lista #2: ")

#Capturar m enterors para L2
L2 = []
for i in range(m):
  num = int(input(f"Introduce el entero {i+1} para L2: "))
  L2.append(num)
print(L2)

#Generar LSUMA

LSUMA = [L1[i]+L2[i] for i in range (m)]
print(f"LSUMA (L1[i]+L2[i]): {LSUMA}")

#Crear LORDENADA
LORDENADA = sorted(L1+L2)
print(f"LORDENADA: {LORDENADA}")

#Crear LUNICA eliminando duplicados de ORDENADA preservando el
# orden de aparición
LUNICA = []
for num in LORDENADA:
  if num not in LUNICA:
    LUNICA.append(num)
print(f"UNICA: {LUNICA}")
```

Opción 1: Dígitos faltantes (1–9)

- Luego lee t valores validados como dígitos 0–9.

- Muestra en un mismo renglón, de menor a mayor, los dígitos entre 1 y 9 que no están en la lista.

**Salida:** una línea con los faltantes separados por espacio (o línea vacía si no falta ninguno).


Opción 2: Serie de Fibonacci en lista
