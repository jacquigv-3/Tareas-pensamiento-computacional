**diagonal_principal:**

- Esta función toma una matriz cuadrada como entrada.

- Crea una lista vacía llamada vector_diagonal.
- Itera sobre cada índice de la matriz (suponiendo que es cuadrada).

- Añade a vector_diagonal los elementos de la diagonal principal de la matriz, es decir, los elementos matriz[x][x].
- Finalmente, retorna esta lista vector_diagonal.

**leer:**

- Toma como entrada el número de renglones (ren) y el número de columnas (col) de la matriz.
- Crea una lista vacía llamada matriz.
- Utiliza dos bucles anidados para llenar la matriz:
  - El bucle exterior itera sobre los renglones.
  - El bucle interior itera sobre las columnas y solicita al usuario que ingrese un valor para cada posición de la matriz.
- Imprime la matriz resultante y la retorna.

**main:**

- Solicita al usuario que ingrese el número de renglones y columnas.
- Verifica si la matriz es cuadrada (si el número de renglones es igual al número de columnas).
- Si no es cuadrada, imprime un mensaje de error.
- Si es cuadrada, llama a la función leer para obtener la matriz, luego llama a diagonal_principal para obtener la diagonal principal, y la imprime.

```python
# -- Función leer ---
def leer(ren, col): # toma como argumento el # de renglones y columnas.
  """
  Esta función arma la matriz -> pide los datos.
  """
  # se crea una lista vacía para guardar las listas de renglones
  matriz = [] # Aquí se guarda la matriz (lista de listas)

  # Este for es para cada renglón (itera para cada renglón)
  for i in range(ren):
    renglon = [] # Lista para los números de este renglón
    print(f"Datos para el Renglón {i}")

    # Este for es para cada columna (los números de adentro)
    for j in range(col):
      # Solicita al usuario el valor para la posición específica [i][j]
      valor = int(input(f"Valor para [{i}][{j}]: "))
      renglon.append(valor) # Mete el número al renglón

    # Añade el renglón (ya lleno) a la matriz principal
    matriz.append(renglon) # Mete el renglón completo a la matriz

  print() # Un espacio para que se vea limpio

 # Imprime la matriz resultante
 print("Tu matriz se ve así:")

  for r in matriz: #Itera sobre cada renglón (lista) en la matriz
    print(r) #Imprime el renglón

  # Da la matriz completa
  return matriz


# --- Función diagonal principal ---
# Esta función toma una matriz como parámetro
def diagonal_principal(matriz):
  """
  Saca solo los números de la diagonal principal.
  """
  # Crea una lista vacía llamada vector_diagonal
  vector_diagonal = []

  # El chiste es que [x][x] te da la diagonal (0,0), (1,1), etc.
  # Usamos len(matriz) para saber cuántas vueltas dar
  for x in range(len(matriz)):
    elemento = matriz[x][x] # Accedemos a [0][0], [1][1], [2][2]...
    vector_diagonal.append(elemento)

  # Finalmente, retorna la lista
  return vector_diagonal

# --- Función main ---
def main_ejercicio1():
  """
  Esta es la función principal del Ejercicio 1.
  """
  print("Ejercicio 1: Diagonal Principal")
  print() # Espacio

# Solicita al usuario que ingrese el número de renglones y columnas
  ren = int(input("Número de renglones: "))
  col = int(input("Número de columnas: "))

  # Checamos si es cuadrada, si no, no se puede
  if ren != col:
    print("Error. La matriz debe ser cuadrada (mismos renglones y columnas).")
  else:

    # Si todo bien, la leemos y calculamos la diagonal
    matriz_leida = leer(ren, col)
    # 2. Llama a diagonal_principal con la matriz leída
    diagonal = diagonal_principal(matriz_leida)

    print() # Espacio

    # 3. Imprime la diagonal principal resultante
    print(f"El vector de la diagonal es: {diagonal}")

# --- Para correr este ejercicio ---
## Llama a la función principal para que el programa inicie
main_ejercicio1()
```

Exercise 2:

**centroMatriz:**

- Esta función toma una matriz como entrada.
- Crea una lista vacía llamada matrizSalida.
- Itera sobre cada renglón de la matriz, excluyendo los bordes (es decir, desde el segundo renglón hasta el penúltimo).
- Dentro de este bucle, crea otra lista vacía llamada vector.
- Itera sobre cada columna de la matriz, excluyendo también los bordes (es decir, desde la segunda columna hasta la penúltima).
- Añade a vector los elementos correspondientes al "centro" de la matriz original.
- Añade vector a matrizSalida.
- Finalmente, retorna matrizSalida.

**leer:**

- Toma como entrada el número de renglones (ren) y el número de columnas (col) de la matriz.
- Crea una lista vacía llamada matriz.
- Utiliza dos bucles anidados para llenar la matriz:
  - El bucle exterior itera sobre los renglones.
  - El bucle interior itera sobre las columnas y solicita al usuario que ingrese un valor para cada posición de la matriz.
- Imprime la matriz resultante y la retorna.

**main:**

- Solicita al usuario que ingrese el número de renglones y columnas.
- Llama a la función leer para obtener la matriz.
- Llama a la función centroMatriz para obtener el "centro" de la matriz.
- Imprime la matriz resultante.

```python
## --- Función leer ---

# Toma como entrada el número de renglones y columnas

def leer(ren, col):
  """
  Esta función arma la matriz pidiéndote los datos.
  """
  matriz = [] # Aquí se guarda la matriz (lista de listas)

  # Este for es para cada renglón
  for i in range(ren):
    renglon = [] # Lista para los números de este renglón
    print(f"Datos para el Renglón {i}")

    # Este for es para cada columna (los números de adentro)
    for j in range(col):
      # y solicita al usuario que ingrese un valor para cada posición
      valor = int(input(f"Valor para [{i}][{j}]: "))
      renglon.append(valor) # Mete el número al renglón

    matriz.append(renglon) # Mete el renglón completo a la matriz

  print() # Un espacio para que se vea limpio

  # Imprime la matriz resultante
  print("Tu matriz se ve así:")
  for r in matriz:
    print(r)

  return matriz

#--- Función centro Matriz ---
# Aqui se toma una matriz como entrada
def centro_matriz(matriz):
  """
  Esta función saca la "carnita" de la matriz, o sea, sin los bordes.
  """
  # Crea una lista vacía llamada matrizSalida
  matrizSalida = []

  # Obtenemos el total de renglones y columnas
  total_renglones = len(matriz)

  # Si la matriz no está vacía, vemos cuántas columnas tiene
  if total_renglones > 0:
    total_columnas = len(matriz[0])
  else:
    total_columnas = 0

  # Empezamos en 1 (el segundo renglón) y terminamos uno ANTES del final
  for i in range(1, total_renglones - 1):
    vector = [] # El nuevo renglón "central"

    # Igual para las columnas, de la 1 a la penúltima
    for j in range(1, total_columnas - 1):
      vector.append(matriz[i][j])

    # Solo agregamos el renglón si de verdad tiene algo
    # (Esto evita agregar [] si la matriz es angosta, ej: 5x2)
    if len(vector) > 0:
      matrizSalida.append(vector)

  return matrizSalida

## ----- Función main ---
def main_ejercicio2():
  """
  Esta es la función principal del Ejercicio 2.
  """
  print("Ejercicio 2: Centro de la Matriz")
  print() # Espacio
  # Solicita al usuario que ingrese el número de renglones y columnas
  ren = int(input("Número de renglones: "))
  col = int(input("Número de columnas: "))

  # Llama a la fucnión leer para obtener la matriz
  matriz_leida = leer(ren, col)

  # Llama a la funcióncentroMatriz para obtener el "centro"
  centro = centro_matriz(matriz_leida)

  print() # Espacio
  print("El centro de la matriz es:")

## Imprime la matriz resultante
  if len(centro) > 0:
    for r in centro:
      print(r)
  else:
    # Esto pasa si la matriz es muy pequeña (ej: 3x3, 2x5, etc.)
    print("[] (La matriz es muy pequeña, no tiene centro)")

# --- Para correr este ejercicio ---
# Llama a la función principal para iniciar el script
main_ejercicio2()
```

Exercise 3:

**leer:**

- Toma como entrada el número de renglones (ren) y columnas (col) de la matriz.
- Crea una lista vacía llamada matriz.
- Utiliza dos bucles anidados para llenar la matriz:
  - El bucle exterior itera sobre los renglones.
  - El bucle interior itera sobre las columnas y solicita al usuario que ingrese un valor para cada posición de la matriz.
- Imprime la matriz resultante y la retorna.

**cambiaRenglones:**

- Pide al usuario que ingrese dos números de renglones que desea intercambiar (restando 1 a cada entrada para ajustarlo a índices de lista basados en cero).
- Itera sobre cada columna de la matriz y realiza el intercambio correspondiente entre los elementos de los dos renglones especificados.
- Finalmente, retorna la matriz con los renglones intercambiados.

**main:**

- Solicita al usuario que ingrese el número de renglones y columnas.
- Llama a la función leer para obtener la matriz.
- Llama a la función cambiaRenglones para intercambiar dos renglones de la matriz.
- Imprime la matriz resultante.

```python
# --- Definición de la Función 'leer' ---

# Define una función llamada 'leer' que acepta dos argumentos: 'ren' (renglones) y 'col' (columnas).
def leer(ren, col):
  """
  Esta función arma la matriz pidiéndote los datos.
  """
  # Crea una lista vacía llamada 'matriz'. Esta será la lista principal que contendrá los renglones.
  matriz = []

  # Inicia un bucle 'for' que se repetirá 'ren' veces (una vez por cada renglón).
  # 'i' tomará los valores 0, 1, 2, ...
  for i in range(ren):
    # Crea una lista vacía llamada 'renglon' para guardar los números de esta fila.
    renglon = []
    # Imprime un mensaje indicando para qué renglón se están pidiendo datos (ej: "Renglón 0").
    print(f"Datos para el Renglón {i}")

    # Inicia un bucle 'for' anidado que se repetirá 'col' veces (para cada columna).
    # 'j' tomará los valores 0, 1, 2, ...
    for j in range(col):
      # Pide al usuario un valor, mostrando la coordenada [i][j], y lo convierte a entero.
      valor = int(input(f"Valor para [{i}][{j}]: "))
      # Agrega el 'valor' recién leído al final de la lista 'renglon'.
      renglon.append(valor)

    # Cuando el bucle de columnas termina, agrega el 'renglon' completo a la 'matriz'.
    matriz.append(renglon)

  print() # Imprime una línea en blanco para separar visualmente.
  print("Tu matriz se ve así:") # Muestra un mensaje al usuario.

  # Inicia un bucle 'for' que recorre cada elemento 'r' (cada renglón) en la 'matriz'.
  for r in matriz:
    print(r) # Imprime el renglón 'r'.

  # Devuelve la 'matriz' completa (la lista de listas) para que pueda ser usada en otra parte.
  return matriz

# --- Definición de la Función 'cambiaRenglones' ---

# Define una función llamada 'cambiaRenglones' que acepta una 'matriz' como argumento.
def cambiaRenglones(matriz):
  """
  Intercambia dos renglones que tú le digas.
  """
  print() # Imprime una línea en blanco para separar.
  print("Vamos a intercambiar renglones.") # Muestra un mensaje al usuario.

  # Pide al usuario el primer número de renglón (ej: 1) y lo convierte a entero.
  r1 = int(input("Qué renglón quieres mover (ej: 1, 2...): "))
  # Pide al usuario el segundo número de renglón (ej: 3) y lo convierte a entero.
  r2 = int(input("Por cuál renglón lo quieres cambiar: "))

  # Resta 1 al número del usuario para obtener el índice real (ej: 1 -> índice 0).
  idx1 = r1 - 1
  # Resta 1 al segundo número para obtener su índice.
  idx2 = r2 - 1

 # -------------

  # 1. Guarda una copia del renglón en 'idx1' (que es una lista) en una variable 'temp'.
  temp = matriz[idx1]

  # 2. Reemplaza el renglón en 'idx1' con el renglón que está en 'idx2'.
  matriz[idx1] = matriz[idx2]

  # 3. Reemplaza el renglón en 'idx2' con lo que habíamos guardado en 'temp'.
  matriz[idx2] = temp

  # Devuelve la 'matriz', que ahora tiene los renglones intercambiados.
  return matriz

# --- Función 'main_ejercicio3' ---

# Define la función principal que ejecutará el programa.
def main_ejercicio3():
  """
  Esta es la función principal del Ejercicio 3.
  """
  # Imprime el título del ejercicio.
  print("Ejercicio 3: Cambiar Renglones")
  print() # Imprime una línea en blanco.

  # Pide al usuario el número de renglones y lo guarda en la variable 'ren'.
  ren = int(input("Número de renglones: "))
  # Pide al usuario el número de columnas y lo guarda en la variable 'col'.
  col = int(input("Número de columnas: "))

  # Llama a la función 'leer', pasándole 'ren' y 'col', y guarda la matriz devuelta en 'matriz_leida'.
  matriz_leida = leer(ren, col)

  # Llama a la función 'cambiaRenglones', pasándole la matriz leída, y guarda el resultado en 'matriz_cambiada'.
  matriz_cambiada = cambiaRenglones(matriz_leida)

  print() # Imprime una línea en blanco.
  print("La matriz con los renglones cambiados:") # Muestra un mensaje.

  # Inicia un bucle 'for' que recorre cada renglón 'r' de la 'matriz_cambiada'.
  for r in matriz_cambiada:
    print(r) # Imprime el renglón actual.

# ---------

# Llama a la función 'main_ejercicio3()' para que el programa empiece a ejecutarse.
main_ejercicio3()
```
