## 1. Diagonal de Rectángulo

Este script calcula la longitud de la diagonal de un rectángulo usando el Teorema de Pitágoras.

```python
#Diagonal de rectángulo

import math

# --- Entrada de Datos ---
#Puntos a tomar en cuenta:
# 1. input() muestra el mensaje dado y lee la respuesta del usuario (siempre como texto).
# 2. float() convierte ese texto en un número decimal (flotante).
# 3. El resultado se guarda en la variable con el nombre asignado (varibel = (valor de varible))

ancho= float(input("Increse el ancho de su rectángulo: ")) #se define la variable, que es un valor decimal y lo ingresa el usuario.
largo= float(input("Increse el largo de su rectángulo: "))

# --- Cálculo de Teorema de Pitágoras ---
# La diagonal (d), el ancho (a) y el largo (l) forman un triángulo rectángulo.
# math.sqrt() calcula la raíz cuadrada de la suma de los cuadrados.
diagonal= math.sqrt((ancho**2)+(largo**2)) #definimos nombre de la varible que engloba la operación a utilizar

# --- SAlida ---
# Utilizamos print para mostrar el resultado.
print(f"Ancho igual a {ancho} largo igual a {largo} por ende la diagonal es {diagonal:.2f}")
```

## 2. Distancia entre dos puntos

Este script calcula la distancia euclidiana entre dos puntos (x1, y1) y (x2, y2) en un plano cartesiano.

```python
#Distancia entre dos puntos

# Se importa math para usar la función de la raíz cuadrada.
import math

# --- Entrada de coordenadas ---

# Coordenada 1
x1 = int(input("ingrese el valor de x1: "))
y1 = int(input("ingrese el valor de y1: "))

# Coordenada 1
x2 = int(input("ingrese el valor de x2: "))
y2 = int(input("ingrese el valor de y2: "))

# --- Cálculo ---
# Esta fórmula también se deriva del Teorema de Pitágoras:
# d = raíz_cuadrada( (delta_x)^2 + (delta_y)^2 )
# donde delta_x es (x2-x1) y delta_y es (y2-y1)

distancia = float(math.sqrt((x2-x1)**2+(y2-y1)**2)) #utilizo .sqrt para sacar raíz

# --- Salida con formato ---
# Se usa un "f-string" (identificado por la 'f' antes de las comillas).
# Esto permite "inyectar" variables directamente en el texto usando llaves {}.
# (:.2f)
  ## ':'  -> Inicia las instrucciones de formato.
  ## '.2' -> Pide redondear a 2 posiciones decimales.
  ## 'f' -> Indica que la variable es un 'float' (decimal).

print(f"distancia= {distancia:.2f}") #en mi string habrá una variable y necesito que solo tenga dos decimales
```

## 3. Litros de pintura

Este código calcula cuántos botes de pintura (L) se necesitan para cubri un área determinada.

```python
#Litros de pintura

# importamos math para poder tener acceso a "math.ceil"
import math # se asegura que math este importado

#--- Entrada de datos ---
area = float(input("Área = ")) # pide ingresar el área a pintat
cobertura_por_litro = float(input("Metros cuadrados que cubre un litro ")) # Pide el rendimiento de la pintura en L.

# --- Cálculo ---
litros_necesarios = area / cobertura_por_litro # calcula la cantidad exacta de L necesarios.

# Como no se puden comprar 5.05L, en esta siguiente varible se redondea el valor obtenido.
litros_completos = math.ceil(litros_necesarios) #math.ceil se usa para redondear un número hacia arriba.

# --- Salida ---
# Con la función print se muestra el número entero de L que se necesitan comprar.
print("Necesitas un total de",litros_completos,"L")
```
